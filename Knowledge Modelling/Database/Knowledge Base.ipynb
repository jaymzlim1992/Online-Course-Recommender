# 2. Aligning Dataset

```python
# Align Header
header = ['title', 'url', 'categories',  'description_short', 'description_long', 'difficulty', 'duration', 'free_option', 'number_of_enroll', 'rating', 'paid_option', 'language', 'subtitle', 'platform', 'provider', 'image_url']
```

```python
# Edx - Remove 'USD' string

edx_edited_data = edx_data[:]
edx_edited_data['Paid Option'] = edx_edited_data['Paid Option'].astype(str).apply(lambda x: x.split(' ')[0])
edx_edited_data['Duration'] = edx_edited_data['Duration'].fillna(0).astype(np.int)
edx_edited_data['Platform'] = edx_edited_data['Platform'].astype(np.int)

# Edx - Align Header
edx_edited_data.columns = header
edx_edited_data.isna().sum()
```

```python
edx_edited_data.head()
```

```python
# Coursera

coursera_edited_data = coursera_data[:]

# Coursera - Align Header
coursera_edited_data.columns = header
coursera_edited_data.isna().sum()
```

```python
coursera_edited_data.head()
```

```python
# Udemy 

udemy_edited_data = udemy_data[:]

# Udemy - Align Header
udemy_edited_data = udemy_edited_data.drop(['Rating', 'Numberofrated'], axis=1)
udemy_edited_data = udemy_edited_data[['Course Name', 'Course URL', 'Categories', 'Short Description',
       'Long Description', 'Difficulty', 'Duration', 'Free Option', 'Numberofenroll', 'Original rating',
       'Paid Option', 'Language', 'Subtitle Language',
       'Platform', 'Provider', 'Image URL']]
udemy_edited_data['Image URL'] = udemy_edited_data['Image URL'].fillna('https://offerage.net/wp-content/uploads/2021/09/new-udemy-1.png')
udemy_edited_data.columns = header
udemy_edited_data.isna().sum()
```

```python
udemy_edited_data.head()
```

# 3. Combine Dataset 

```python
dataset = pd.concat([edx_edited_data, coursera_edited_data, udemy_edited_data]).reset_index(drop=True)
```

```python
print(dataset.shape)
dataset.head()
```

```python
dataset.isna().sum()
```

```python
# Calculate Popularity Index

avg_rating = dataset['rating'][dataset['rating']>0].mean()
pop_index = dataset['number_of_enroll']/dataset['number_of_enroll'].max() * \
            ((dataset['rating']!=0) * (dataset['rating'] / avg_rating) + \
             (dataset['rating']==0) * 1)
dataset['popularity_index'] = pop_index
print(avg_rating)
pop_index.shape
```

```python
print(dataset.shape)
dataset.head()
```

## > Build Database

```python
# Load Course Data from pickle

dataset_db = pd.read_pickle(pkl_file)
dataset_header = ['index']+ list(dataset_db.columns)
dataset_db['index'] = dataset_db.index + 1
dataset_db = dataset_db[dataset_header]


course_header = ['courseID',
                 'title',
                 'url',
                 'categories',
                 'description_short',
                 'description_long',
                 'difficulty',
                 'duration',
                 'free_option',
                 'number_of_enroll',
                 'rating',
                 'paid_option',
                 'language',
                 'subtitle',
                 'platform',
                 'provider',
                 'image_url',
                 'popularity_index']

dataset_db.columns = course_header
dataset_db.head()
```

```python
# SQL QUERY - CREATE TABLE

course_create_table = ''' CREATE TABLE IF NOT EXISTS course 
                        (
                        courseID INTEGER NOT NULL PRIMARY KEY,
                        title TEXT NOT NULL,
                        url TEXT NOT NULL,
                        categories TEXT,
                        description_short TEXT,
                        description_long TEXT,
                        difficulty INTEGER NOT NULL,
                        duration INTEGER NOT NULL,
                        free_option INTEGER,
                        number_of_enroll INTEGER,
                        rating REAL,
                        paid_option TEXT,
                        language TEXT,
                        subtitle TEXT,
                        platform INTEGER NOT NULL,
                        provider TEXT,
                        image_url TEXT, 
                        popularity_index REAL
                        );
                        '''


user_create_table = ''' CREATE TABLE IF NOT EXISTS user
                        (
                        userID INTEGER NOT NULL PRIMARY KEY, 
                        name varchar(255) NOT NULL,
                        username varchar(255) NOT NULL UNIQUE, 
                        password varchar(255) NOT NULL
                        );
                        '''

favourite_create_table = ''' CREATE TABLE IF NOT EXISTS favourite
                             (
                             userID INTEGER NOT NULL,
                             courseID INTEGER NOT NULL,
                             FOREIGN KEY (userID) REFERENCES user (userID),
                             FOREIGN KEY (courseID) REFERENCES course (courseID)
                             );
                             '''

query_create_table = ''' CREATE TABLE IF NOT EXISTS query
                         (
                         queryID INTEGER PRIMARY KEY, 
                         query_count INTEGER NOT NULL, 
                         userID INTEGER NOT NULL,
                         query_time DATETIME DEFAULT (CURRENT_TIMESTAMP),
                         query_text TEXT, 
                         query_duration INTEGER,
                         query_difficulty INTEGER, 
                         query_free_option INTEGER,
                         FOREIGN KEY (userID) REFERENCES user (userID)
                         );
                         '''

recom_create_table = ''' CREATE TABLE IF NOT EXISTS recommendation
                         (
                         userID INTEGER NOT NULL, 
                         query_count INTEGER NOT NULL, 
                         ranking INTEGER NOT NULL, 
                         courseID INTEGER NOT NULL,
                         FOREIGN KEY (userID) REFERENCES user (userID),
                         FOREIGN KEY (query_count) REFERENCES query (query_count),
                         FOREIGN KEY (courseID) REFERENCES course (courseID)
                         );
                         '''
```

```python
# EXECUTE CREATE TABLE

filename = 'app_database.db'

conn = sqlite3.connect(filename)
conn.execute(course_create_table)
conn.execute(user_create_table)
conn.execute(favourite_create_table)
conn.execute(query_create_table)
conn.execute(recom_create_table)

conn.close()
```
